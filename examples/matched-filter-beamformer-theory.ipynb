# Spatially Band-Limited Matched Filter Beamformer

### Spherical Harmonics Expansion

##### Frequency domain

The sound field on a spherical surface can be represented as a series of spherical harmonics,

\begin{align}
S(\boldsymbol{x}, \omega)
=& \sum_{n=0}^{\infty} \sum_{m=-n}^{n}
\mathring{S}_{nm}(r, \omega)
Y_{nm}(\hat{\boldsymbol{x}}),
\end{align}

where
- $\boldsymbol{x} = (r, \theta, \phi)$ :
position vector in the spherical coordinate system
($\theta$: colatitude, $\phi$: azimuth)
- $\hat{\boldsymbol{x}} = \frac{\boldsymbol{x}}{\Vert\boldsymbol{x}\Vert}
= (1, \theta, \phi)$ : unit vector pointing $\boldsymbol{x}$,
therefore $\boldsymbol{x} = r \hat{\boldsymbol{x}}$
- $Y_{nm}(\hat{\boldsymbol{x}})$ :
spherical harmonic of order $n$ and degree $m$
evaluated at $\hat{\boldsymbol{x}}$
- $\omega=2\pi f$ : angular frequency
- $S(\boldsymbol{x}, \omega)$ : sound field
- $\mathring{S}_{nm}(r, \omega)$ : expansion coefficient of the sound field
- $c$ : speed of sound

For the sound field of a single plane wave,
$\mathring{S}_{nm} = 4\pi i^{n} j_{n}(\tfrac{\omega}{c}r)
Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw})$,

\begin{align}
e^{i \tfrac{\omega}{c}r \langle \boldsymbol{n}_\text{pw}, \hat{\boldsymbol{x}} \rangle}
=& \sum_{n=0}^{\infty}
4 \pi i^{n} j_{n}(\tfrac{\omega}{c}r)
\sum_{m=-n}^{n}
Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw})
Y_{nm}(\hat{\boldsymbol{x}}) \\
=& \sum_{n=0}^{\infty}
(2n + 1) i^{n} j_{n}(\tfrac{\omega}{c}r)
P_{n}(\langle \boldsymbol{n}_\text{pw}, \hat{\boldsymbol{x}} \rangle),
\end{align}

where

- $\boldsymbol{n}_\text{pw} = (1, \theta_\text{pw}, \phi_\text{pw})$ :
unit vector pointing the origin of the plane wave
- $j_{n}(\tfrac{\omega}{c}r)$ : spherical Bessel function of the first kind
- $P_{n}(\cdot)$ : Legendre polynomial
- $\langle \cdot, \cdot \rangle$ : scalar product of the two vectors
- $(\cdot)^{\ast}$ : complex conjugate
- $\sum_{m=-n}^{n} Y_{nm}(\boldsymbol{n}_1) Y_{nm}^\ast(\boldsymbol{n}_2)
= \frac{2n + 1}{4\pi} P_{n}(\langle\boldsymbol{n}_1\boldsymbol{n}_2\rangle)$
is exploited in the last equality

##### Time-Domain

The time-domain representation of the spherical harmonics expansion reads

\begin{align}
\delta\left(t + \tfrac{r}{c} \langle \boldsymbol{n}_\text{pw},
\hat{\boldsymbol{x}} \rangle \right)
=& \frac{c}{2r}
\sum_{n=0}^{\infty} 4 \pi
(-1)^{n} \tilde{P}_{n}(\tfrac{c}{r}t)
\sum_{m=-n}^{n}
Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw})
Y_{nm}(\hat{\boldsymbol{x}}) \\
=& \frac{c}{2r}
\sum_{n=0}^{\infty} (2n+1)
(-1)^{n} \tilde{P}_{n}(\tfrac{c}{r}t) \:
P_{n}(\langle \boldsymbol{n}_\text{pw},
\hat{\boldsymbol{x}} \rangle) \\
=& \frac{c}{2r}
\sum_{n=0}^{\infty} (2n+1)
\tilde{P}_{n}(-\tfrac{c}{r}t) \:
P_{n}(\langle \boldsymbol{n}_\text{pw},
\hat{\boldsymbol{x}} \rangle)
\end{align}

where
- the inverse Fourier transform of $j_{n}(\tfrac{\omega}{c}r)$ is used

\begin{align}
\frac{1}{2\pi}
\int_{-\infty}^{\infty} j_{n}(\tfrac{\omega}{c}r) \:
e^{i\omega t} \text{d}\omega =
i^{n} \tfrac{c}{2r} \tilde{P}_{n}(\tfrac{c}{r}t)
\end{align}

- $\tilde{P}_{n}(\tfrac{c}{r}t)$ : a piecewise continuous function
which coincides with $P_{n}(\tfrac{c}{r}t)$ within
$[-\tfrac{r}{c}, \tfrac{r}{c}]$ and zero elsewhere,

\begin{equation}
\tilde{P}_{n}(\tfrac{c}{r}t) \equiv
\begin{cases}
P_{n}(\tfrac{c}{r}t) & \text{for $t \in (-\frac{r}{c}, \frac{r}{c})$} \\
\frac{1}{2} P_{n}(\tfrac{c}{r}t) & \text{for $t = \pm\frac{r}{c}$} \\
0 & \text{otherwise}
\end{cases}
\end{equation}

- $(-1)^{n} P_{n}(x) = P_{n}(-x)$ is exploited in the last equality

In the remainder, both $\tilde{P}_{n}(\cdot)$ and $P_{n}(\cdot)$
will be denoted by $P_{n}(\cdot)$ for notational convenience.

The time-domain expansion coefficient of a plane wave thus reads

$$\mathring{s}_{nm}(r, t) = \frac{2\pi c}{r} (-1)^{n}
P_{n}(\tfrac{c}{r}t) Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw}).$$

##### Legendre series

The time-domain expression can be considered as a Legendre series,

$$
\delta\left(t + \tfrac{r}{c} \langle \boldsymbol{n}_\text{pw}, \hat{\boldsymbol{x}} \rangle \right)
= \sum_{n=0}^{\infty} \alpha_{n} P_{n}(\tfrac{c}{r}t),
\quad \text{for $t \in [-\tfrac{r}{c}, \tfrac{r}{c}]$}
$$

with the expansion coefficient

$$
\alpha_{n} = \frac{c}{2r} (-1)^{n} (2n + 1) P_{n}(\langle \boldsymbol{n}_\text{pw}, \hat{\boldsymbol{x}} \rangle)
$$

Note the orthogonality of the Legnedre polynomials,

$$
\int_{-1}^{1} P_m(\mu) P_n(\mu) \text{d}\mu
= \frac{2}{2n + 1} \delta_{mn}
\quad \xrightarrow[]{\mu \rightarrow \tfrac{c}{r}t} \quad
\frac{c}{r}
\int_{-\frac{r}{c}}^{\frac{r}{c}} P_m(\tfrac{c}{r}t) P_n(\tfrac{c}{r}t) \: \text{d}t
= \frac{2}{2n + 1} \delta_{mn}
$$

### Matched Filter Beamformer

##### Frequency Domain
(Please refer to [Rafaely, 2015, Ch. 5] for a thorough introduction.)

In a typical modal beamforming, the captured sound field on a
sphere $\mathring{S}_{nm}(r=R, \omega)$ is pre-equalized
by a modal weight $B_{n}(R, \omega)$,

$$
\mathring{S}_{nm}(R, \omega) \times \frac{1}{B_{n}(R, \omega)},
$$

which is determined by the physical structure of the array.
For an open sphere (free-field within $r \le R$), for instance,
$B_{n}(R, \omega) = 4 \pi i^{n} j_{n}(\tfrac{\omega}{c}R)$
which is the radial part of $\mathring{S}_{nm}(R, \omega)$ for
a plane wave.
An additional modal weight $\mathring{D}_{nm}(\boldsymbol{n}_l, \omega)$
is applied to obtain a desired beam pattern,

\begin{align}
\mathring{V}_{nm}(\boldsymbol{n}_l, \omega)
=& \mathring{S}_{nm}(R, \omega) \times \frac{1}{B_{n}(R, \omega)}
\times \mathring{D}_{nm}(\boldsymbol{n}_l, \omega) \\
=& \mathring{S}_{nm}(R, \omega) \:
\mathring{W}_{nm}(\boldsymbol{n}_l, \omega),
\end{align}

where
* $\boldsymbol{n}_l$ : look direction of the beamformer
* $\mathring{V}_{nm}(\boldsymbol{n}_l, \omega)$ : beamformer output
* $\mathring{W}_{nm}(\boldsymbol{n}_l, \omega) =
\mathring{D}_{nm}(\boldsymbol{n}_l, \omega)/ B_{n}(R, \omega)$ :
modal weight of the beamformer

For an axis-symmetric beamformer, which is aiming for in the following,
the modal weight can be separated into radial and angular parts,

$$
\mathring{W}_{nm}(\boldsymbol{n}_l, \omega) =
\frac{D_{n}(\omega)}{B_{n}(R, \omega)}
Y_{nm}(\boldsymbol{n}_l)
$$

For a **matched filter beamformer**, the radial part is 

$$D_{n}(\omega) = |B_{n}(R, \omega)|^2
= B_{n}(R, \omega) \: B_{n}^{\ast}(R, \omega)$$

and therefore

$$
\mathring{W}_{nm}(\boldsymbol{n}_l, \omega) =
B_{n}^{\ast}(R, \omega) Y_{nm}(\boldsymbol{n}_l).
$$

Since the term $B_{n}(R, \omega)$ in the denominator is canceled out,
no pre-equalization is performed here.
The array structure thus affects the beamformer output.

For an open sphere, the modal weight reads

$$
\mathring{W}_{nm}^\circ(\boldsymbol{n}_l, \omega)
= 4 \pi i^{-n} j_{n}^{\ast}(\tfrac{\omega}{c}R) Y_{nm}(\boldsymbol{n}_l),
$$

If this is applied to a single incident plane wave,
the frequency response is characterized by
the square of the spherical Bessel function,

\begin{align}
\mathring{V}_{nm}(\boldsymbol{n}_l, \omega)
=& 4 \pi i^{n} j_{n}(\tfrac{\omega}{c}R)
Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw})
\times 4 \pi i^{-n} j_{n}^{\ast}(\tfrac{\omega}{c}R)
Y_{nm}(\boldsymbol{n}_l) \\
=& 16 \pi^{2} |j_{n}(\tfrac{\omega}{c}R)|^2
Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw})Y_{nm}(\boldsymbol{n}_l).
\end{align}

Finally, the beam pattern is given as the series expansion,

\begin{align}
V(\boldsymbol{n}_{l}, \omega)
=& \sum_{n=0}^{\infty} \sum_{m=-n}^{n}
\mathring{V}_{nm}(\boldsymbol{n}_{l}, \omega) \\
=& 16 \pi^{2} \sum_{n=0}^{\infty} |j_{n}(\tfrac{\omega}{c}R)|^2
\sum_{m=-n}^{n} Y_{nm}(\boldsymbol{n}_\text{pw})Y_{nm}^{\ast}(\boldsymbol{n}_l) \\
=& 4 \pi \sum_{n=0}^{\infty} (2n + 1) |j_{n}(\tfrac{\omega}{c}R)|^2
P_{n}(\langle\boldsymbol{n}_\text{pw}, \boldsymbol{n}_l\rangle).
\end{align}

In practice, the infinite summation is approximated by

$$
V(\boldsymbol{n}_{l}, \omega)
\approx 4 \pi \sum_{n=0}^{N} (2n + 1) |j_{n}(\tfrac{\omega}{c}R)|^2
P_{n}(\langle\boldsymbol{n}_\text{pw}, \boldsymbol{n}_l\rangle).
$$

The maximum modal order $N$ is called the modal or spatial
bandwidth of the beamformer, hence the name
**spatially band-limited** matched filter beamformer.


##### Simulation

In the following script, the beam pattern `V` is computed
from the analytic expression given above.
The microphones are assumed to have a continuous distribution
on an open sphere, and no spatial sampling is considered.
The beam patterns are computed for different look directions
on the horizontal plane `azi_l`.
The frequency response on the on-axis
($\boldsymbol{n}_l = \boldsymbol{n}_\text{pw}$)
is simulated for varying spatial bandwidth.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import fftconvolve as conv, freqz
from scipy.special import spherical_jn, eval_legendre as legendre
%matplotlib inline

def db(x, power=False):
    with np.errstate(divide='ignore'):
        return 10 if power else 20 * np.log10(np.abs(x))
```

```python
# Horizontal beam pattern of a spatially matched beamformer
c = 343  # speed of sound
R = 0.25  # array radius
max_order = 7  # modal bandwidth

fmin, fmax, num_f = 80, 15000, 3000
f = np.logspace(np.log10(fmin), np.log10(fmax), num=num_f)
kr = 2 * np.pi * f / c * R
fN = c * max_order / 2 / np.pi / R  # critical frequency

azi_pw, colat_pw = 0, np.pi / 2  # incident plane wave
azi_l, colat_l = np.linspace(-np.pi, np.pi, num=200), np.pi / 2  # look dir.
ctheta = np.cos(colat_pw) * np.cos(colat_l)\
         + np.sin(colat_pw) * np.sin(colat_l) * np.cos(azi_pw - azi_l)

V = np.sum(np.stack([4 * np.pi * (2 * n + 1) # beam pattern
                     * spherical_jn(n, kr[:, np.newaxis])**2
                     * legendre(n, ctheta)
                     for n in range(max_order + 1)]), axis=0)

Vn = np.stack([4 * np.pi * (2 * n + 1) # on-axis responses
              * spherical_jn(n, kr)**2
              * legendre(n, 1)
              for n in range(max_order + 1)])


fig, ax = plt.subplots(figsize=(12, 4), ncols=2)

im = ax[0].pcolormesh(np.rad2deg(azi_l), f, db(V), vmin=-60)
if fN < fmax: ax[0].plot(np.rad2deg(azi_l), fN * np.ones_like(azi_l), 'w--')
ax[0].set_xlabel('Azimuth / deg')
ax[0].set_ylabel('Frequency / Hz')
ax[0].set_yscale('log')
fig.colorbar(im, ax=ax[0]).set_label('dB')
ax[0].set_title('Beam Pattern (N={})'.format(max_order))

[ax[1].semilogx(f, db(np.sum(Vn[:n+1, :], axis=0)),
                label='{}'.format(n))
                for n in range(max_order + 1)]
ax[1].plot(fN, db(4 * np.pi), color='k', marker='v', ms=12, alpha=0.3)
ax[1].set_xlim([fmin, fmax])
ax[1].set_ylim(ymin=-60)
ax[1].grid(which='both')
ax[1].set_xlabel('Frequency / Hz')
ax[1].set_ylabel('Magnitude / dB')
ax[1].legend(loc='lower left', title='modal bandwidth')
ax[1].set_title('On-axis Magnitude Responses');

print('R = {R} m, N = {max_order}, f_N = {fN}'
     .format(R=R, max_order=max_order, fN=np.round(fN)))
```

For a given modal bandwidth $N$, a critical frequency is defined as
$f_N = \frac{c N}{2 \pi R}$.
It is indicated by the dashed line in the left figure,
and by a triangle in the right one.
For $f < f_N$, the main lobe gets narrower
as the frequency get closer to $f_N$.
This shows the equivalence of the matched filter beamformer
to the delay-and-sum beamformer in the frequency range
where the modal band-limitation has negligible effect.
For $f > f_N$, the main lobe width remain almost constant
with minor variations.

As shown in the right figure, the magnitude response
for the on-axis direction is almost flat up to $f_N$.
The transfer function is real-valued.
Above $f_N$, the magnitude response exhibits
a decay of about -12 dB/Oct.
This agrees with the large arguement approximation
of the spherical Bessel function.

##### Time Domain

In this section, the matched filter beamformer is formulated in the time domain.
The corresponding beamformer coefficient is given as
the inverse Fourier transform of
$\mathring{W}_{nm}(\boldsymbol{n}_l, \omega)$ (open sphere),

\begin{align}
\mathring{w}_{nm}(\boldsymbol{n}_l, t)
=& \mathcal{F}^{-1}\{\mathring{W}_{nm}(\boldsymbol{n}_l, \omega)\} \\
=& 4 \pi i^{-n} Y_{nm}(\boldsymbol{n}_l) \:
\mathcal{F}^{-1}\{j_{n}^{\ast}(\tfrac{\omega}{c}R)\} \\
=& ??? \\
=& \tfrac{2 \pi c}{R} (-1)^{n} Y_{nm}(\boldsymbol{n}_l) \: P_{n}(-\tfrac{c}{R}t) \\
=& \tfrac{2 \pi c}{R} Y_{nm}(\boldsymbol{n}_l) \: P_{n}(\tfrac{c}{R}t).
\end{align}

The beamformer operation is the convolution of the latter
with the modal coefficient of the captured signal,

$$
\mathring{v}_{nm}(\boldsymbol{n}_l, t)
= \tfrac{2 \pi c}{R}
Y_{nm}(\boldsymbol{n}_l)
P_{n}(\tfrac{c}{R}t) \ast_{t} \mathring{s}_{nm}(R, t) ,
$$

where

* $\mathring{v}_{nm}(\boldsymbol{n}_l, t)$ : modal coefficient of the beamformer output
* $\mathring{s}_{nm}(R, t)$ : modal coefficient of the captured signal
* $\ast_{t}$ : time-domain convolution

The corresponding output thus reads

\begin{align}
v(\boldsymbol{n}_l, t)
=& \sum_{n=0}^{\infty} \sum_{m=-n}^{n}
\mathring{v}_{nm}(\boldsymbol{n}_l,t)\\
=& \tfrac{2\pi c}{R}
\sum_{n=0}^{\infty} \sum_{m=-n}^{n}
P_{n}(\tfrac{c}{R}t) \ast_{t} \mathring{s}_{nm}(R, t)
Y_{nm}(\boldsymbol{n}_{l}) \\
=& \tfrac{2 \pi c}{R}
\sum_{n=0}^{N}
P_{n}(\tfrac{c}{R}t) \ast_{t}
\sum_{m=-n}^{n}
\mathring{s}_{nm}(R, t) Y_{nm}(\boldsymbol{n}_{l})
\end{align}

Notice that $N+1$ rather than $(N+1)^2$ convolutions are required
as the time-dependent term $P_{n}(\tfrac{c}{R}t)$ only depends on $n$.

For a single plane wave, where
$\mathring{s}_{nm}(R, t) = \frac{2 \pi c}{R} (-1)^{n} P_{n}(\tfrac{c}{R}t)
Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw})$,
the beamformer output reads

\begin{align}
v(\boldsymbol{n}_l, t)
=& (\tfrac{2\pi c}{R})^2
\sum_{n=0}^{\infty}
(-1)^{n}
P_{n}(\tfrac{c}{R}t) \ast_{t} P_{n}(\tfrac{c}{R}t)
\sum_{m=-n}^{n}
Y_{nm}^{\ast}(\boldsymbol{n}_\text{pw}) Y_{nm}(\boldsymbol{n}_l) \\
=& \pi (\tfrac{c}{R})^2
\sum_{n=0}^{\infty}
(2n + 1) (-1)^{n}
P_{n}(\langle\boldsymbol{n}_\text{pw}, \boldsymbol{n}_l\rangle)
P_{n}(\tfrac{c}{R}t) \ast_{t} P_{n}(\tfrac{c}{R}t) \\
=& ???
\end{align}

Notice that the impulse response for each $n$ is characterized by
the convolution of two Legendre polynomial of the same order,

$$P_{n}(\tfrac{c}{R}t) \ast_{t} P_{n}(\tfrac{c}{R}t).$$

According to [Hale and Townsend, 2014], the latter consists of two
piecewise polynomials,

\begin{align}
v_{n}(t) =
\begin{cases}
v_{n}^\text{left}(t), & t \in [-\tfrac{2R}{c}, 0] \\
v_{n}^\text{right}(t), & t \in (0, \tfrac{2R}{c}] \\
0, & \text{otherwise}
\end{cases}
\end{align}

each of which can be expressed as a Legendre series of $(2n+1)$-th order,

\begin{align}
v_{n}^\text{left}(t) &= \sum_{k=0}^{2n+1} \gamma_k^\text{left} P_{k}(t + \tfrac{R}{c}), & t \in [-\tfrac{2R}{c}, 0] \\
v_{n}^\text{left}(t) &= \sum_{k=0}^{2n+1} \gamma_k^\text{right} P_{k}(t - \tfrac{R}{c}), & t \in (0, \tfrac{2R}{c}].
\end{align}

```python
# Impulse responses of a spatially matched beamformer
c = 343  # speed of sound
fs = 44100 * 4  # sampling frequency
R = 0.25  # array radius

Lh = int(np.floor(R / c * fs))
t = np.arange(-2 * Lh, 2 * Lh + 1) / fs
crt = c / R * np.arange(-Lh, Lh + 1) / fs
dt = 1 / fs

max_order = 7  # modal bandwidth

azi_pw, colat_pw = 0, 0.5 * np.pi  # incident plane wave
azi_l, colat_l = np.linspace(-np.pi, np.pi, num=200), 0.5 * np.pi  # look dir.
ctheta = np.cos(colat_pw) * np.cos(colat_l)\
         + np.sin(colat_pw) * np.sin(colat_l) * np.cos(azi_pw - azi_l)

v = np.sum(np.stack([np.pi * (c / R)**2 * dt  # impulse response
                     * (2 * n + 1) * (-1)**n
                     * conv(legendre(n, crt[:, np.newaxis]),
                            legendre(n, crt[:, np.newaxis]))
                     * legendre(n, ctheta)
                     for n in range(max_order + 1)]), axis=0)

vn = np.stack([np.pi * (c / R)** 2 * dt  # on-axis impulse response
               * (2 * n + 1) * (-1)**n
               * conv(legendre(n, crt),
                      legendre(n, crt))
               * legendre(n, 1)
               for n in range(max_order + 1)])

fig, ax = plt.subplots(figsize=(12, 4), ncols=2)

im = ax[0].pcolormesh(np.rad2deg(azi_l), t * 1000, db(v), vmin=20)
ax[0].set_xlabel('Azimuth / deg')
ax[0].set_ylabel('Time / ms')
cbar = fig.colorbar(im, ax=ax[0]).set_label('dB')
ax[0].set_title('Impulse Responses (N={})'.format(max_order));

[ax[1].plot(t * 1000, np.sum(vn[:n+1, :], axis=0).T,
            label='{}'.format(n))
            for n in range(max_order + 1)]
ax[1].grid(which='both')
ax[1].set_xlabel('Time / ms')
ax[1].legend(loc='upper right', title='modal bandwidth')
ax[1].set_title('On-Axis Impulse Responses');

print('R = {R} m, N = {max_order}, R/c = {roverc} ms'
     .format(R=R, max_order=max_order, roverc=R/c*1000))
```

The impulse responses of the beamformer exhibit symmetry around $t=0$ and
have nonzero support for $t\in[-\tfrac{2R}{c}, \tfrac{2R}{c}]$.
The energy distribution in the time-domain depends on
the angle between $\boldsymbol{n}_\text{pw}$ and $\boldsymbol{n}_l$.
As the look direction gets closer to the incident plane wave,
more energy is concentrated around $t=0$, thus more peaky.
This is more so for higher modal bandwidth as shown in the right figure.
The impulse responses decay towards both ends with oscillation.
For modal bandwidth $n$, $2n$ zero crossings ($n$ on each side)
are observed excluding the ones at $t=\pm\tfrac{2R}{c}$.
